```python
import serial
import time


arduino = serial.Serial('COM3', 9600, timeout=1)
time.sleep(2)  # Wait for connection to establish

# Send message
message = "Hello Arduino!"
arduino.write(message.encode())
print(f"Sent: {message}")

# Close connection
arduino.close()
```
