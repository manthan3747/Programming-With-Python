```python
# Post Lab Exercise 2: Fill missing values with mean of column
import pandas as pd
import numpy as np

print("Post Lab Exercise 2: Filling Missing Values")
print("-" * 40)

# creating sample data with some missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [30, np.nan, 35, 28, np.nan],
    'Salary': [50000, 60000, np.nan, 55000, 58000]
}

df_missing = pd.DataFrame(data)

print("Original DataFrame with missing values:")
print(df_missing)
print()

# checking which values are missing
print("Missing values:")
print(df_missing.isnull().sum())
print()

# filling missing values in Age column with mean
mean_age = df_missing['Age'].mean()
df_missing['Age'].fillna(mean_age, inplace=True)

print(f"Mean age calculated: {mean_age}")
print()

# filling missing values in Salary column with mean
mean_salary = df_missing['Salary'].mean()
df_missing['Salary'].fillna(mean_salary, inplace=True)

print(f"Mean salary calculated: {mean_salary}")
print()

print("DataFrame after filling missing values:")
print(df_missing)
print()

# verify no missing values left
print("Missing values after filling:")
print(df_missing.isnull().sum())
```
