```python

import numpy as np
import matplotlib.pyplot as plt

def unit_step_ztransform():
    """
    Compute and analyze the Z-transform of unit step function
    U(z) = 1/(1 - z^-1)
    """

    print("Z-Transform of Unit Step Function")
    
    # For unit step function: U(z) = 1/(1 - z^-1)
    # In standard form: numerator = [1], denominator = [1, -1]
    
    numerator = [1]  # Coefficient of numerator
    denominator = [1, -1]  # Coefficients: 1 - 1*z^-1
    
    print("\nTransfer Function:")
    print("U(z) = 1 / (1 - z^-1)")
    print("\nNumerator coefficients:", numerator)
    print("Denominator coefficients:", denominator)
    
    # Find poles (roots of denominator)
    poles = np.roots(denominator)
    print("\nPoles of the system:", poles)
    
    # Check stability
    print("\n Stability Analysis")
    stable = True
    for i, pole in enumerate(poles):
        magnitude = np.abs(pole)
        print(f"Pole {i+1}: {pole:.4f}, Magnitude: {magnitude:.4f}")
        if magnitude >= 1:
            stable = False
    
    if stable:
        print("\nResult: System is STABLE")
        print("(All poles lie inside the unit circle)")
    else:
        print("\nResult: System is UNSTABLE")
        print("(One or more poles lie outside or on the unit circle)")
    
    # Plot poles in z-plane
    plt.figure(figsize=(8, 8))
    
    # Draw unit circle
    theta = np.linspace(0, 2*np.pi, 100)
    x_circle = np.cos(theta)
    y_circle = np.sin(theta)
    plt.plot(x_circle, y_circle, 'b--', label='Unit Circle')
    
    # Plot poles
    for pole in poles:
        plt.plot(np.real(pole), np.imag(pole), 'rx', markersize=15, 
                markeredgewidth=3, label='Pole')
    
    plt.axhline(y=0, color='k', linewidth=0.5)
    plt.axvline(x=0, color='k', linewidth=0.5)
    plt.grid(True, alpha=0.3)
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.title('Pole-Zero Plot for Unit Step Function')
    plt.axis('equal')
    plt.xlim([-1.5, 1.5])
    plt.ylim([-1.5, 1.5])
    plt.legend()
    plt.show()
    
    return stable

# Run the analysis
if __name__ == "__main__":
    stability = unit_step_ztransform()
    print("Analysis Complete!")
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("POST LAB EXERCISE - LTI SYSTEM ANALYSIS USING Z-TRANSFORM")

# post lab 1
print("\n\n")
print("EXERCISE 1: Z-TRANSFORM OF UNIT STEP FUNCTION")

def exercise1_unit_step():
    """
    Z-transform of unit step: U(z) = 1/(1 - z^-1)
    """
    print("\nUnit Step Function: u[n] = 1 for n >= 0")
    print("Z-Transform: U(z) = 1/(1 - z^-1)")
    
    # Coefficients
    numerator = [1]
    denominator = [1, -1]
    
    print(f"\nNumerator: {numerator}")
    print(f"Denominator: {denominator}")
    
    # Calculate poles
    poles = np.roots(denominator)
    print(f"\nPoles: {poles}")
    
    # Stability check
    print("\n--- Checking Stability ---")
    for i, pole in enumerate(poles):
        mag = np.abs(pole)
        print(f"Pole {i+1}: {pole:.4f}, |pole| = {mag:.4f}")
        
        if mag < 1:
            print(f"   Inside unit circle")
        elif mag == 1:
            print(f"   On unit circle (Marginally stable)")
        else:
            print(f"   Outside unit circle")
    
    # Check if stable
    stable = all(np.abs(p) < 1 for p in poles)
    
    if stable:
        print("\n STABLE: All poles inside unit circle")
    else:
        print("\n UNSTABLE: Poles on or outside unit circle")
    
    return poles, stable

# Run Exercise 1
poles1, stable1 = exercise1_unit_step()

# post lab 2
print("\n\n")
print("EXERCISE 2: ANALYZE H(z) = 0.5(z-0.7)(z-0.9)/(z-0.6)(z-0.4)")

def exercise2_system_hz():
    """
    Analyze H(z) = 0.5(z-0.7)(z-0.9) / (z-0.6)(z-0.4)
    """
    # System parameters
    zeros = np.array([0.7, 0.9])
    poles = np.array([0.6, 0.4])
    gain = 0.5
    
    print(f"\nGain: {gain}")
    print(f"Zeros: {zeros}")
    print(f"Poles: {poles}")
    
    # Stability Analysis
    print("\n--- Checking Stability ---")
    for i, pole in enumerate(poles):
        mag = np.abs(pole)
        print(f"Pole {i+1}: {pole:.4f}, |pole| = {mag:.4f}")
        
        if mag < 1:
            print(f"   Inside unit circle")
        elif mag == 1:
            print(f"   On unit circle")
        else:
            print(f"   Outside unit circle")
    
    # Check stability
    stable = all(np.abs(p) < 1 for p in poles)
    
    if stable:
        print("\n STABLE: All poles inside unit circle")
    else:
        print("\n UNSTABLE: Poles on or outside unit circle")
    
    # Other properties
    print("\n--- Other Properties ---")
    print("Causality: CAUSAL (numerator degree <= denominator degree)")
    print("Time Invariance: TIME-INVARIANT (constant coefficients)")
    
    return zeros, poles, stable


zeros2, poles2, stable2 = exercise2_system_hz()

print("\n\n")

print("GENERATING POLE-ZERO PLOTS...")


# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Function to draw unit circle and axes
def setup_plot(ax, title):
    theta = np.linspace(0, 2*np.pi, 200)
    x_circle = np.cos(theta)
    y_circle = np.sin(theta)
    ax.plot(x_circle, y_circle, 'b-', linewidth=2, label='Unit Circle')
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('Real Part', fontsize=11)
    ax.set_ylabel('Imaginary Part', fontsize=11)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.axis('equal')
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])

# Plot 1: Unit Step
setup_plot(ax1, 'Exercise 1: Unit Step Function')
for pole in poles1:
    ax1.plot(np.real(pole), np.imag(pole), 'rx', markersize=15, 
            markeredgewidth=3, label='Pole')
ax1.legend(loc='upper right')
status1 = "UNSTABLE" if not stable1 else "STABLE"
ax1.text(0.5, -1.3, f'Status: {status1}', ha='center', 
        fontsize=11, fontweight='bold',
        color='red' if not stable1 else 'green')

# Plot 2: System H(z)
setup_plot(ax2, 'Exercise 2: System H(z)')
for zero in zeros2:
    ax2.plot(np.real(zero), np.imag(zero), 'go', markersize=12, 
            markeredgewidth=2, markeredgecolor='darkgreen', label='Zero')
for pole in poles2:
    ax2.plot(np.real(pole), np.imag(pole), 'rx', markersize=15, 
            markeredgewidth=3, label='Pole')
# Remove duplicate labels
handles, labels = ax2.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax2.legend(by_label.values(), by_label.keys(), loc='upper right')
status2 = "UNSTABLE" if not stable2 else "STABLE"
ax2.text(0.5, -1.3, f'Status: {status2}', ha='center', 
        fontsize=11, fontweight='bold',
        color='red' if not stable2 else 'green')

plt.tight_layout()
plt.show()

print("\n\n")
print("SUMMARY OF RESULTS")
print(f"\nExercise 1 - Unit Step Function:")
print(f"  Status: {'UNSTABLE' if not stable1 else 'STABLE'}")
print(f"  Poles: {poles1}")

print(f"\nExercise 2 - System H(z):")
print(f"  Status: {'UNSTABLE' if not stable2 else 'STABLE'}")
print(f"  Zeros: {zeros2}")
print(f"  Poles: {poles2}")

print("\n" + "="*70)
print("POST LAB EXERCISE COMPLETED!")
```
